The table above is to be read in the following manner: Given MyChoice = {A, B, C}, and the Price being behind one in {A, B, C}, then the probabilty of each door being opened by the official is either {0, 0.5, 1}.

Now computing P(ContainsPrize | OpenedByOfficial, MyChoice) = P(OpenedByOfficial | ContainsPrice, MyChoice) * P(ContainsPrice) / P(OpenedByOfficial | MyChoice). Where P(OpenedByOfficial | MyChoice) = sum over P(OpenedByOfficial | ContainsPrize, MyChoice)*P(ContainsPrice)

Eg: P(A |B, A) = P(B | A, A)*P(A) / P(B | A) <br>
P(B | A) = P(B |Â A, A) * P(A) + P(B | C, A) * P(C) = 1/2 * 1/3 + 1 * 1/3 = 1/2 <br>
=> P(A | B, A) = (1/2 * 1/3) / (1/2) = 1/3
